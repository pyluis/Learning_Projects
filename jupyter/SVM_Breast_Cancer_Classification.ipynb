# Step 1: import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
#concatenate data and target arrays
cancer_df = pd.DataFrame(np.c_[cancer['data'],cancer['target']], 
                         columns=np.append(cancer['feature_names'],'target'))
```

# Step 3: Visualize Data

# Step 4: Model Training

```python
X = cancer_df.drop(['target'], axis= 1)
```

```python
X
```

```python
y = cancer_df.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC(gamma='auto')
```

```python
svc_model.fit(X_train, y_train)
```

# Step 5: Model Testing and Evaluation

```python
y_predict = svc_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

# Step 6: Improving the Model

```python
min_train = X_train.min()
```

```python
min_train
```

```python
range_train = (X_train - min_train).max()
```

```python
range_train
```

```python
x_train_scaled = (X_train - min_train)/range_train
```

```python
x_train_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue=y_train)
```

```python
sns.scatterplot(x = x_train_scaled['mean area'], y = x_train_scaled['mean smoothness'], hue =y_train )
```

```python
min_test = X_test.min()
```

```python
range_test = (X_test - min_test).max()
```

```python
x_test_scaled = (X_test - min_test)/range_test
```

```python
svc_model = SVC(gamma='auto')
svc_model.fit(x_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(x_test_scaled)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
print(classification_report(y_test,y_predict))
```

```python
param_grid = {
    'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001], 'kernel':['rbf']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(gamma='auto'),param_grid,refit=True, verbose = 4,error_score='raise')
```

```python
grid.fit(x_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid_predict = grid.predict(x_test_scaled)
```

```python
cm = confusion_matrix(y_test,grid_predict)
```

```python
sns.heatmap(cm, annot=True)
```
